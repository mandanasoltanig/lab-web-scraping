```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
BASE_URL = "https://books.toscrape.com/"
FANTASY_URL = "https://books.toscrape.com/catalogue/category/books/fantasy_19/index.html"

rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}
```

```python
def get_soup(url):
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(url, headers=headers, timeout=20)
    response.raise_for_status()
    response.encoding = "utf-8"
    return BeautifulSoup(response.text, "html.parser")
```

```python
def get_fantasy_book_links():
    links = []
    page_url = FANTASY_URL

    while page_url:
        print("Visiting:", page_url)
        soup = get_soup(page_url)

        books = soup.find_all("article", class_="product_pod")

        for book in books:
            a_tag = book.find("h3").find("a")
            relative_link = a_tag["href"]
            full_link = urljoin(page_url, relative_link)
            links.append(full_link)

        next_button = soup.find("li", class_="next")
        if next_button:
            next_page = next_button.find("a")["href"]
            page_url = urljoin(page_url, next_page)
        else:
            page_url = None

    return links
```

```python
fantasy_links = get_fantasy_book_links()
print("Fantasy links found:", len(fantasy_links))
print(fantasy_links[:5])
```

```python
def scrape_book_details(book_url):
    try:
        soup = get_soup(book_url)

        title_tag = soup.find("h1")
        title = title_tag.text.strip() if title_tag else "Unknown"

        rating = None
        rating_tag = soup.find("p", class_="star-rating")
        if rating_tag:
            rating_classes = rating_tag.get("class", [])
            for item in rating_classes:
                if item in rating_map:
                    rating = rating_map[item]
                    break

        table = soup.find("table")
        if table is None:
            return None

        rows = table.find_all("tr")
        info = {}

        for row in rows:
            th = row.find("th")
            td = row.find("td")
            if th and td:
                info[th.text.strip()] = td.text.strip()

        upc = info.get("UPC", "")
        price_text = info.get("Price (incl. tax)", "£0.00")
        price = float(price_text.replace("Â", "").replace("£", "").strip())
        availability = info.get("Availability", "")

        genre = "Fantasy"

        description = ""
        desc_header = soup.find("div", id="product_description")
        if desc_header:
            desc_paragraph = desc_header.find_next_sibling("p")
            if desc_paragraph:
                description = desc_paragraph.text.strip()

        return {
            "UPC": upc,
            "Title": title,
            "Price (£)": price,
            "Rating": rating,
            "Genre": genre,
            "Availability": availability,
            "Description": description
        }

    except Exception as e:
        print("Error scraping:", book_url)
        print(e)
        return None
```

```python
def scrape_books(min_rating, max_price):
    data = []

    fantasy_links = get_fantasy_book_links()
    print("Total Fantasy books found:", len(fantasy_links))

    for i, link in enumerate(fantasy_links, start=1):
        print(f"Scraping Fantasy book {i}/{len(fantasy_links)}")

        book = scrape_book_details(link)

        if book is None:
            continue

        if (
            book["Rating"] is not None and
            book["Rating"] >= min_rating and
            book["Price (£)"] <= max_price
        ):
            data.append(book)

    return pd.DataFrame(data, columns=[
        "UPC",
        "Title",
        "Price (£)",
        "Rating",
        "Genre",
        "Availability",
        "Description"
    ])
```

```python
df = scrape_books(4, 20)
print(df[["Title", "Price (£)", "Rating", "Genre"]])
print("Number of matching books:", len(df))
```

```python
df.to_csv("fantasy_books_filtered.csv", index=False)
print("Saved as fantasy_books_filtered.csv")
```
